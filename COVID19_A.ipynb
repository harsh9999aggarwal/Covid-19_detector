```python
import pandas as pd
```

```python
# create the data for +ve samples( github )

FILE_PATH="covid-chestxray_github/metadata.csv"
IMAGES_PATH="covid-chestxray_github/images"
```

```python
df=pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
import os
import shutil
```

```python
TARGET_DIR="Dataset/Covid"

if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("covid folder created")
```

```python
cnt=0

for (i,row) in df.iterrows():
    if row ["finding"]=="COVID-19" and row["view"]=="PA":
        filename=row["filename"]
        image_path=os.path.join(IMAGES_PATH,filename)
        image_copy_path = os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
        print("moving image",cnt)
        cnt+=1
print(cnt)
```

```python
#sampling of images from  kaggle

import random
KAGGLE_FILE_PATH="ChestX-Ray_Kaggle/chest_xray/train/NORMAL"
TARGET_NORMAL_DIR="Dataset/Normal"

if not os.path.exists(TARGET_NORMAL_DIR):
    os.mkdir(TARGET_NORMAL_DIR)
    print("covid folder created")
```

```python
images_names=os.listdir(KAGGLE_FILE_PATH)
```

```python
random.shuffle(images_names)
```

```python
for i in range(176):
    image_name=images_names[i]
    
    image_path=os.path.join(KAGGLE_FILE_PATH,image_name)
    target_path=os.path.join(TARGET_NORMAL_DIR,image_name)
    shutil.copy2(image_path,target_path)
    print("copying image" , i)
```
